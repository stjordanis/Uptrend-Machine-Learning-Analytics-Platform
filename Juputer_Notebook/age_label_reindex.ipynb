```python
# Ddd dependencies
import pandas as pd
import numpy as np
```

```python
# Load csv files
r_data = 'prob.csv'
```

```python
# Read Data File and store into Pandas Data Frames
r_df = pd.read_csv(r_data)
r_df.head()
```

```python
# Feature engineer gender column
r_df = r_df.replace(18, 1)
r_df = r_df.replace(25, 1)
r_df = r_df.replace(35, 2)
r_df = r_df.replace(45, 2)
r_df = r_df.replace(55, 3)

r_df.head()
```

```python
# Rename column
r_df.rename(columns={'all_live_theater/concerts/dance-attended_last_12_months': 'all_live','social_media_[social_media_user]': 'social_media','supermarket_goer_last_4_weeks': 'supermarket'}, inplace=True)
r_df.head()
```

```python
# Save clean data to csv
r_df.to_csv('r_data_18_35_55.csv', index = False)
```
