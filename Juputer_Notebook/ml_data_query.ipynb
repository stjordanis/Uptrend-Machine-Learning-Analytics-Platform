```python
# Import dependencies
import sqlite3
import pandas as pd
import numpy as np
from numpy import nan
```

```python
# Connect to database
conn = sqlite3.connect('database.sqlite')
```

# Query for Male Data

```python
# Read sqlite query results into a pandas DataFrame
male_df = pd.read_sql_query("""
SELECT 
*
FROM east_central_table

where gender = 'Male' and category not in ('totals') and year not in (2017)

union

SELECT 
*
FROM north_east_table

where gender = 'Male' and category not in ('totals') and year not in (2017)

union

SELECT 
*
FROM pacific_table

where gender = 'Male' and category not in ('totals') and year not in (2017)

union

SELECT 
*
FROM south_table

where gender = 'Male' and category not in ('totals') and year not in (2017)

union

SELECT 
*
FROM south_east_table

where gender = 'Male' and category not in ('totals') and year not in (2017)

union

SELECT 
*
FROM south_west_table

where gender = 'Male' and category not in ('totals') and year not in (2017)

union

SELECT 
*
FROM west_central_table

where gender = 'Male' and category not in ('totals') and year not in (2017)

""",conn)
    
# Drop na and first 3 columns
male_df = male_df.drop(male_df.columns[[0,2]], axis=1)

male_df.head()
```

```python
# Feature engineer category column
male_df['category'] = male_df['category'].map({
'video_game': '1',
'tv': '2',
'radio': '3',
'magazine': '4',
'movie_goers': '5',
'supermarket_goer_last_4_weeks': '6',
'tablet_owner': '7',
'all_live_theater/concerts/dance-attended_last_12_months': '8',
'social_media_[social_media_user]': '9',
'digital_music': '10',
'streaming_video': '11'
})

# Feature engineer age column
male_df['age'] = male_df['age'].map({
'18-24': '18',
'25-34': '25',
'35-44': '35',
'45-54': '45',
'55+': '55'})

# Feature engineer division column
male_df['division'] = male_df['division'].map({
'east_central': '1',
'north_east': '2',
'pacific': '3',
'south': '4',
'south_east': '5',
'south_west': '6',
'west_central': '7'})

male_df.head()
```
