# Clean, Pivot and Merge Data for East Central Division 2013 - 2017

```python
# create clean data frame
clean_data_ec = pd.DataFrame(columns=['category','gender','age','year','division','population'])
clean_data_ec
```

```python
# Loop thru 2013 - 2017 data and create combined data frame
for year in [2013, 2014, 2015, 2016, 2017]:
    #print(year) 
    
    raw_data_ec = pd.read_csv("../raw_data/east_central_data/Movie Purchaser Behavior EC {}.csv".format(year),skiprows=1)
#raw_data_ec

    # Drop na and first 3 columns
    raw_data_ec = raw_data_ec.dropna().drop(raw_data_ec.columns[[1, 2, 3]], axis=1)
#raw_data_ec

    # Strip first column
    raw_data_ec.iloc[:,0] = raw_data_ec.iloc[:,0].str.strip()
#raw_data_ec

    # Replace whitespace with "_" in category column 
    raw_data_ec.iloc[:,0] = raw_data_ec.iloc[:,0].str.replace(' ', '_')
#raw_data_ec

    # Lowercase string in first column 
    raw_data_ec.iloc[:,0] = raw_data_ec.iloc[:,0].str.lower()
#raw_data_ec

    # Create a list for category name
    cat_list = list(raw_data_ec.iloc[:,0])    
    cat_list_final = []    
    for i in cat_list:
        cat_list_final += [i]*len(['Male', 'Female'])*len(['18-24', '25-34', '35-44', '45-54', '55+'])
#print(cat_list_final)

    # Create a list for population in each category
    pop_list = []
    for index, row in raw_data_ec.iterrows():
        pop_list += [int(row[5])] +  [int(row[8])]\
                    + [int(row[11])] + [int(row[14])]\
                    + [int(row[17])]+ [int(row[6])]\
                    + [int(row[9])] + [int(row[12])]\
                    + [int(row[15])]+ [int(row[18])]
    #print(pop_list)

    # create yearly data frame
    yearly_data_ec = pd.DataFrame(columns=['category','gender','age','year','division','population'])

    # Move values into clean_ec_data data frame
    yearly_data_ec["category"] = cat_list_final
    yearly_data_ec['gender'] = (['Male']*len(['18-24', '25-34', '35-44', '45-54', '55+'])+['Female']*len(['18-24', '25-34', '35-44', '45-54', '55+']))*len(cat_list)
    yearly_data_ec['age'] = ['18-24', '25-34', '35-44', '45-54', '55+']*len(['Male', 'Female'])*len(cat_list)
    yearly_data_ec['year'] = [year]*len(cat_list)*len(['Male', 'Female'])*len(['18-24', '25-34', '35-44', '45-54', '55+'])
    yearly_data_ec['division'] = ['east_central']*len(cat_list)*len(['Male', 'Female'])*len(['18-24', '25-34', '35-44', '45-54', '55+'])
    yearly_data_ec['population'] = pop_list
    
    # Append yearly datafram generated from loop into clean data frame 
    clean_data_ec = pd.concat([clean_data_ec,yearly_data_ec], ignore_index=True)

    print ("{} data finished".format(year))
    
clean_data_ec.tail(10)
```

```python
# Create a new df that will contain all region all year share (pop of category / total pop)
prob_df_all = pd.DataFrame(columns=['category','gender','age','year','division','population',"share"])
```
